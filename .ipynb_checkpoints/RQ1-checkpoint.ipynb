```python
import pandas as pd
```

```python
df = pd.DataFrame()

pickup = pd.Series()
zoneId = pd.Series()
df['pickup'] = pd.Series()
df['LocationID'] = pd.Series()
buffer = 10000
# , nrows=buffer*100
for month in ['01', '02', '03', '04', '05', '06']:
    for chunk in pd.read_csv("yellow_tripdata_2018-"+month+".csv", chunksize=buffer, nrows=buffer*10, usecols=['tpep_pickup_datetime', 'PULocationID']):
        pickup = pickup.append(chunk['tpep_pickup_datetime'])
        # I have to store the location id of the pickup, concatenating at the end of a list
        zoneId = zoneId.append(chunk['PULocationID'])


df = df.append(pd.DataFrame({'pickup': pd.to_datetime(
    pickup), 'LocationID': zoneId}), ignore_index=True)
# this dataset have a map between the LocationID and Borough
zone_lookup = pd.read_csv("taxi_zone_lookup.csv")
df = df.merge(zone_lookup[['LocationID', 'Borough']], how='inner', on='LocationID').fillna(
    "")  # Inner join between the interested columns and the previous dataset
df = df[['pickup', 'Borough']]
df = df.loc[df['pickup'].dt.year > 2017]
```

```python
df2 = pd.DataFrame()
tmp = pd.Series()
months = [31, 28, 31, 30, 31, 30]
for borough in set(df['Borough']):
    for i in range(1, max(df['pickup'].dt.month)+1):
        tmpdf = df.loc[(df['pickup'].dt.month == i) & (
            df['Borough'] == borough)]['pickup']
        if len(tmpdf) != 0:
            tmp = tmp.append(
                pd.Series(tmpdf.count()/months[i-1]), ignore_index=True)
    df2[borough] = tmp
    del tmp
    tmp = pd.Series()
df2.rename(index={0: 'Jan', 1: 'Feb', 2: 'Mar',
                  3: 'Apr', 4: 'May', 5: 'Jun'}, inplace=True)
df2 = df2.fillna(0)
df2
```

```python
df2.plot.bar(figsize=(20,20),subplots=True,layout=(5,2))
```
